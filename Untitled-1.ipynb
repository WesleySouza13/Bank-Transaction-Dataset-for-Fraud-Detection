```python
def calcular_parte_dia(Hora):
    if 0 <= Hora < 6:
        return 'Madrugada'
    elif 6 < Hora < 12:
        return 'Manha'
    elif 12 < Hora < 18:
        return 'Tarde'
    else: 
        return 'Noite'
        
```

```python
data['periodo_dia'] = data['Hora'].apply(calcular_parte_dia)
```

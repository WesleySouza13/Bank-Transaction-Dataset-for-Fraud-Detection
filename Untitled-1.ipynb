```python
#leitura dos dados
data = pd.read_csv('bank_transactions_data_2.csv')
data
```

TransactionAmount: Valor monetário de cada transação, variando de pequenas despesas cotidianas a compras maiores.
TransactionType: Campo categórico que indica transações de 'Crédito' ou 'Débito'
Localização: Localização geográfica da transação, representada pelos nomes das cidades dos EUA.
AccountBalance: Saldo na conta pós-transação, com correlações lógicas baseadas no tipo e valor da transação.
Canal: Canal pelo qual a transação foi realizada (ex.: Online, ATM, Agência).
CustomerAge: Idade do titular da conta, com agrupamentos lógicos baseados na ocupação. 
CustomerOccupation: Ocupação do titular da conta (por exemplo, médico, engenheiro, estudante, aposentado), refletindo padrões de renda.
TransactionDuration: Duração da transação em segundos, variando de acordo com o tipo de transação.
LoginAttempts: Número de tentativas de login antes da transação, com valores mais altos indicando possíveis anomalias.

o fator que devemos considerar é o horario da transação 
transações fraudulentas tendem a ser feitas fora do horario abtual, ou seja, de a noite ou na madrugada 


```python
#coletando dados relacionados a horario da transaçao 
data['TransactionDate'] = pd.to_datetime(data['TransactionDate'])
```

```python
data['TransactionDate']
```

```python
#criando nova coluna hora 
data['Hora'] = data['TransactionDate'].dt.hour
```

```python
data['Hora']
```

```python
#funçao para classificaçao da parte do dia referente a hora 
def calcular_parte_dia(Hora):
    if 0 <= Hora < 6:
        return 'Madrugada'
    elif 6 < Hora < 12:
        return 'Manha'
    elif 12 < Hora < 18:
        return 'Tarde'
    else: 
        return 'Noite'
        
```

```python
#aplicaçao da coluna hora para a nova coluna 'periodo dia'
data['periodo_dia'] = data['Hora'].apply(calcular_parte_dia)
```
